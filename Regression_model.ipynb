```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv('suzuki_stock_data.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.index = df['Date']
df.drop('Date', axis = 1, inplace = True)
```

```python
df['returns'] = df['Close'].pct_change()
```

```python
df.head()
```

```python
df = df.dropna(subset = ['returns'])
```

```python
df.head()
```

```python
X = df.drop('returns', axis = 1)
y = df['returns']
```

```python
split = int(len(df)* 0.9)
X_train = X[:split]
X_test = X[split:]
y_train = y[:split]
y_test = y[split:]
```

```python
model_rfr = RandomForestRegressor(n_estimators = 100, random_state = 42)
model_rfr.fit(X_train, y_train)
```

```python
y_pred = model_rfr.predict(X_test)
```

```python
r2_score = model_rfr.score(X_test, y_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
r2_score
```

```python
mse
```

```python
mae
```

```python
model_xg = XGBRegressor(n_estimators = 100, random_state = 42)
model_xg.fit(X_train, y_train)
```

```python
y_pred_xg = model_xg.predict(X_test)
```

```python
r2_score = model_xg.score(X_test, y_test)
mse = mean_squared_error(y_test, y_pred_xg)
mae = mean_absolute_error(y_test, y_pred_xg)
```

```python
r2_score
```

```python
model_svr = SVR(kernel = 'rbf')
model_svr.fit(X_train, y_train)
```

```python
y_pred_svr = model_svr.predict(X_test)
r2_score = model_svr.score(X_test, y_test)
mse = mean_squared_error(y_test, y_pred_svr)
mae = mean_absolute_error(y_test, y_pred_svr)
```

```python
r2_score
```

```python
df['lag1'] = df['Close'].shift(1)
df['lag2'] = df['Close'].shift(2)
df['rolling_mean3']= df['Close'].rolling(window = 3).mean()
```

```python
df.head()
```

```python
df.dropna(subset = ['lag1', 'lag2', 'rolling_mean3'], inplace = True)
```

```python
df.head()
```

```python
X_new = df[['lag1', 'lag2', 'rolling_mean3']]
y_new = df['returns']
```

```python
X_train_new = X_new[:split]
X_test_new = X_new[split:]
y_train_new = y_new[:split]
y_test_new = y_new[split:]
```

```python
model_rfr.fit(X_train_new, y_train_new)
```

```python
r2_score = model_rfr.score(X_test_new, y_test_new)
print(r2_score)
```

```python
model = {
    'Random Forest Regressor': RandomForestRegressor(n_estimators = 100, random_state = 42),
    'XGBoost Regressor': XGBRegressor(n_estimators = 100, learning_rate = 0.1, max_depth = 6, random_state = 42),
    'Support Vector Regressor': SVR(kernel = 'rbf')
}
for key, model in model.items():
    model.fit(X_train_new, y_train_new)
    r2_score = model.score(X_test_new, y_test_new)
    print(f'{key} R2 Score: {r2_score}')
```

```python
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
```

```python
tscv = TimeSeriesSplit(n_splits = 5)

para_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 6, 9],
    'subsample': [0.8,1.0]
}
grid_search = GridSearchCV(estimator = XGBRegressor(random_state = 42), param_grid = para_grid, cv = tscv, scoring = 'r2', n_jobs = -1)
```

```python
grid_search.fit(X_train_new, y_train_new)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
y_pred_gscv = grid_search.best_estimator_.predict(X_test_new)
r2_score = grid_search.best_estimator_.score(X_test_new, y_test_new)
print(f'Grid Search CV XGBoost R2 Score: {r2_score}')
```
